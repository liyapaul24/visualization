```python
#Import all required libraries
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
#Defining tips dataset,
tips = sns.load_dataset("tips")
print(tips)
```

```python
#line graph
sns.lineplot(data=tips, # choose our dataset
             x='total_bill', # define our x variable
             y='tip') # define our y variable
sns.set_style('whitegrid')
```

```python
#Create a line plot with the tips dataset
tipgraph = sns.lineplot(data=tips,
                        x='total_bill',
                        y='tip')
tipgraph.set(title='Tips vs. Total Bill',
             xlabel='Total Bill ($)',
             ylabel='Tip Amount ($)')
```

```python
# Create a line plot using the seaborn library with the 'tips' dataset
fig = plt.subplots(figsize=(10, 3))
tipgraph = sns.lineplot(data=tips,
                        x='total_bill',
                        y='tip',
                        color = 'hotpink',
                        linestyle = '--',
                        linewidth = 3,
                        marker = 'o', # Add circular markers at each data point
                        markerfacecolor = 'indigo')
tipgraph.set(title='Tips vs. Total Bill',
             xlabel='Total Bill ($)',
             ylabel='Tip Amount ($)')
```

```python
# Create a scatter plot using the seaborn library with the 'tips' dataset
tipgraph = sns.scatterplot(data=tips, 
                           x='total_bill',
                           y="tip",
                           style = "time", 
                           hue ="day", # Differentiate points by the 'day' column (will color points based on the day)
                           palette = ["purple","hotpink", "deepskyblue","yellowgreen"])
tipgraph.set(title='Tips vs. Total Bill',
                xlabel='Total Bill ($)',
                ylabel="Tip Amount ($)")
```

```python
#The pairplot generates a grid of scatterplots showing pairwise relationships between all numerical columns in the tips dataset.
sns.pairplot(data=tips,
             hue='day')
```

```python
#Create a relational plot (scatter plot) with multiple subplots based on the 'day' column
daysplot = sns.relplot(data=tips,
                       x="total_bill",
                       y="tip",
                       hue="sex",
                       col="day",
                       kind="scatter",
                       col_wrap=2)
```

```python
#Creating data arrays with names and values for plotting with plotly
import plotly.graph_objects as go # 'go' is 'graph objects
x1 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
```

```python
graph = go.Figure() # Initialize a new empty figure for the bar chart.

graph.add_trace(go.Bar(x=x1, y=y1)) # Add a bar trace to the figure with x as pirate names and y as their corresponding scores.

graph.update_layout(title="Pirate Scores",
                    xaxis_title="Pirates",
                    yaxis_title="Score")
```

```python
graph = go.Figure() # Initialize a new empty figure for the scatter plot.
graph.add_trace(go.Scatter(x=x1, y=y1, mode='markers', # Add a scatter trace with points (markers) for pirates and their scores.
    marker=dict(
    size=15, # point size
    color='hotpink', # point colour
    opacity=1, # point transparency/alpha
    line=dict(width=5, color='purple') # point outline
    ))
)
graph.update_layout(title='Interactive Pirate Plot',
                    xaxis_title='Pirates',
                    yaxis_title='Scores',
                    width=500, height=500)
```

```python
from wordcloud import WordCloud

df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv", # Read a CSV file from the provided URL
on_bad_lines='skip') # Skip any rows in the CSV that contain malformed data.
df   # Display the DataFrame.
```

```python
# join all our text from each row from our quote column into a string
text = " ".join(each for each in df.quote)

# generate our wordcloud image
wordcloud = WordCloud(background_color="white",
                      colormap = 'inferno').generate(text)

# use matplotlib syntax to put our image in a figure
fig, ax = plt.subplots(figsize=(7, 3))
ax.imshow(wordcloud, # remember 'imshow' from when we added pictures to our matplotlib axes
          interpolation='bilinear') # this line helps smooth our image

ax.axis("off")
```

```python
# Import functions for creating 2-set Venn diagrams and customizations in Matplotlib
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted 
```

```python
#define our sets
A = set(["apple", "banana", "watermelon"])
B = set(["pumpkin", "blueberry", "apple", "key lime"])

diagram = venn2_unweighted([A, B],  # Create a 2-set Venn diagram for sets A and B, with unweighted counts.
                           set_labels = ('Fruits', 'Pies'),
                           set_colors=("blue", "red"),
                           alpha=0.5) # Set the transparency (alpha) of the diagram to 50%.
diagram.get_label_by_id("10").set_text("\n".join(A - B))
diagram.get_label_by_id("11").set_text("\n".join(A & B))
diagram.get_label_by_id("01").set_text("\n".join(B - A))
```
