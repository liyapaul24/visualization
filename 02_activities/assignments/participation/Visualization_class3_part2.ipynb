```python
#Import all required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy","Zoro","Nami","Usopp","Sanji"]) # Create an array of character names for x2
y2 = np.array([110, 180, 240, 99, 220]) # Create an array of corresponding y-values for the character names
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,nrows=1,figsize=(7, 3)) # Create a figure with two subplots side by side (1 row, 2 columns) with a size of 7x3 inches
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
```

Activity

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,nrows=1,figsize=(7, 3)) 
#ax1.scatter(x1,y1)
ax1.plot(x1,y1,
           marker='*',
           color = 'green',
           linestyle = '--',
           linewidth=2,
           markeredgecolor = '#FA9359',
           markerfacecolor = '#000000')

ax2.bar(x2,y2,
        color=['#4CAF50', '#FF5722', '#2196F3', '#9C27B0', '#FFC107'],  # Custom colors for each bar
               edgecolor='black',  # Black border around bars
               linewidth=1.5) 
plt.xticks(rotation=45, ha='right')
```

```python

# Create a figure with a 2x2 grid of subplots, where 'ax1', 'ax2', and 'ax3' are specified
# 'ax3' spans both columns in the second row
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],['ax2','ax3']],
figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
```

```python
#adding very large x axis titles
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],['ax2','ax3']],figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
```

```python
# The 'layout="constrained"' ensures that the subplots are automatically adjusted for optimal spacing.
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],['ax2','ax3']],figsize=(7, 4), layout = "constrained")
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
```
