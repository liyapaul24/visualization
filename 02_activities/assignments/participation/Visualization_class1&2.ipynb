```python
#Import all required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
#Initialize the values to generate graph
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

#Generate scatter plot
fig, ax = plt.subplots(figsize =(5,3))
ax.scatter(x,y)
```

```python
#Generate bar chart
fig, ax = plt.subplots(figsize=(5,3)) #chart size in inches
ax.bar(x,y)
```

```python
#Generate line graph 
fig, ax = plt.subplots(figsize=(5,3)) #chart size in inches
ax.plot(x,y)
```

```python
#Generate histogram
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y)
```

```python
#Plot a line graph 
fig, ax = plt.subplots(figsize=(5,3)) 
ax.plot(x,y)

#Set the title and axis labels
ax.set_title('Total growth over time')
ax.set_ylabel('Total Growth')
ax.set_xlabel('Years Since Start')
```

```python
#Define custom font styles
font1 ={'family':'sans-serif','color':'blue','size':20}
font2 ={'family':'monospace','color':'green','size':14}

#plot a line graph 
fig, ax = plt.subplots(figsize =(5,3))
ax.plot(x,y)

#apply the custom fonts to the title and axis labels
ax.set_title('Total growth over time', fontdict=font1)
ax.set_ylabel('Total Growth', fontdict= font2)
ax.set_xlabel('Years Since Start', fontdict=font2)
```

```python
#plot a scatter plot of 'x' vs 'y' with star-shaped markers and indigo color
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y,
           marker='*',
           color = 'indigo')
```

```python
# plot a line graph of 'x' vs 'y' with star-shaped markers, indigo color, dashed line style, and a linewidth of 2
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
           marker='*',
           color = 'indigo',
           linestyle = '--',
           linewidth=2)
```

```python
# plot a line graph of 'x' vs 'y' with star-shaped markers (indigo color), dashed line style, linewidth of 2, and custom marker colors

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
           marker='*',
           color = 'indigo',
           linestyle = '--',
           linewidth=2,
           markeredgecolor = '#FA9359',
           markerfacecolor = '#000000')
ax.grid() #display a grid on the plot
```

```python
# plot a line graph of 'x' vs 'y' with star-shaped markers (indigo color), dashed line style, linewidth of 2, and custom marker colors
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
           marker='*',
           color = 'indigo',
           linestyle = '--',
           linewidth=2,
           markeredgecolor = '#FA9359',
           markerfacecolor = '#000000')
# Add a grid to the plot for the y-axis only, with green color, a linewidth of 1, and a dashed line style
ax.grid(axis = 'y', color = "green", linewidth = 1, linestyle = '--')
```

```python
# Sample DataFrame creation (you should replace this with your actual DataFrame)
data = {
    'type': np.random.choice(['Group1', 'Group2'], size=50),
    'value': np.random.randint(0, 100, size=50)
}
df = pd.DataFrame(data)

# Get group names and define colors
group_name = df['type'].unique()
colors = ['green', 'orange']

# Create the plot
fig, ax = plt.subplots(figsize=(5, 3))

# Plot the histograms
for i, group in enumerate(group_name):
    df[df['type'] == group]['value'].hist(ax=ax, 
                                          bins=12, 
                                          edgecolor='gray', 
                                          color=colors[i], 
                                          alpha=0.6, # Adding alpha for transparency
                                          label=group)

# Add a legend
ax.legend(title='Group')

# Add a title and axis labels
ax.set_title('Distribution of 2 Different Groups')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')

# Show the plot
plt.show()
```
