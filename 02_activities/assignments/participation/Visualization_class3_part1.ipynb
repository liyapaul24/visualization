```python
#Import all required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
#Initialize the values to generate graph
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
#Generate line graph 
fig, ax = plt.subplots(figsize=(5,3)) 
ax.plot(x,y1)
ax.plot(x,y2)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" ) # Plot the data for Person 1
ax.plot(x,y2, label = "Person 2" ) # Plot the data for Person 2
ax.legend(loc='lower right') # Add a legend to the plot, placing it at the lower right corner
```

```python
# Create and display a line plot comparing two datasets with a legend.
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
# Add a legend to the plot, positioning it in the upper-left corner and adjusting its location outside the plot area
ax.legend(loc="upper left",
bbox_to_anchor =(1, 1))
```

```python
# Create and display a line plot comparing two datasets with a legend.
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.text(10, 95,"This value is important!") #Adds a text annotation at the specified coordinates 
```

```python
# Create and display a line plot comparing two datasets with a legend.
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.text(25, 90,"This value is important!",
        ha='center', #alignment of text
        color = 'red', #modify font colour
        size = 20) #modify font size
```

```python
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10]) # Set axis limits from 0 to 10 for both x and y axes

ax.text(1, 5,". Data:(1, 5)", transform=ax.transData ) # Add text at data coordinates (1, 5) using data space
ax.text(0.5, 0.1,". Axes:(0.5, 0.1)", transform=ax.transAxes ) # Add text at axes coordinates (0.5, 0.1) using axes space
ax.text(0.2, 0.2,". Figure:(0.2, 0.2)", transform=fig.transFigure ) # Add text at figure coordinates (0.2, 0.2) using figure space
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),  # Add an annotation with an arrow at (10, 95), pointing to (20, 94), with a black arrow
            arrowprops=dict(facecolor='black', shrink=0.05))
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),  # Add an annotation with a hotpink wedge-shaped arrow from (10, 95) to (20, 94)
            arrowprops=dict(facecolor='hotpink', arrowstyle = 'wedge'))
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.yaxis.set_major_locator(plt.NullLocator()) #locator to remove ticks and lables
ax.xaxis.set_major_formatter(plt.NullFormatter()) #formatter to remove only lables
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right') # Add a legend in the lower-right corner
ax.xaxis.set_major_locator(plt.MaxNLocator(3)) # Limits the x-axis to display no more than 3 major ticks.
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # Set the x-axis to display major ticks at intervals of 5 units
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc="lower right")
plt.xticks(rotation=45, ha='right') # Rotate the x-axis labels by 45 degrees and align them to the right
```

```python
font1 = {'family':'serif','color':'indigo'} # Define custom font properties for the x-axis label
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc="lower right")
plt.xticks(rotation=45, ha='right')
plt.xlabel("Shiny New X Axis!", fontsize = 18, fontdict = font1) # Set the x-axis label with custom font properties and a font size of 18
```
